```python
import psycopg2
import pandas as pd
import matplotlib.pyplot as plt
from dotenv import load_dotenv
import warnings
warnings.filterwarnings('ignore')
import os
```

```python
load_dotenv()

conn = psycopg2.connect(
    host=os.getenv('DB_HOST'),
    database=os.getenv('DB_NAME'),
    user=os.getenv('DB_USER'),
    password=os.getenv('DBPASSWORD')
)
    
```

```python
print(conn.status)
```

```python
#leitura de dados

df_prices = pd.read_sql("""
    SELECT dmc.name, dcmd.captured_at, fct.price_usd,
           fct.market_cap, fct.volume_24h, fct.price_change_pct
    FROM fact_prices as fct
    INNER JOIN dim_coins as dmc ON fct.coin_id = dmc.coin_id
    INNER JOIN dim_date as dcmd ON fct.date_id = dcmd.date_id
""", conn)
```

```python
df_prices
```

```python
filtered_df = df_prices[df_prices['name'] == 'Bitcoin']

filtered_df.plot(x='captured_at', y='price_usd', kind='line')
plt.title('Evolucao de Preco')
plt.xlabel('Data')
plt.ylabel('Price USD')
plt.show()

```

```python
grouped= df_prices.groupby('name').last()
```

```python
grouped
```

```python
grouped['price_change_pct'].plot.bar(color= 'steelblue', figsize=(8,5))
plt.title('Variacao de preco')
plt.xlabel('name')
plt.ylabel('price_changepct')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
